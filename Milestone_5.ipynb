# Milestone 5 – Probability
This notebook contains sample code for each section of Milestone 5.

## Section A: Setup & Load Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load your dataset
# Replace with the actual filename
df = pd.read_csv('insurance.csv')
df.head()
```

## Section B: Treat Dataset as Sample Space

```python
N = len(df)
print('Total observations:', N)
```

## Section C: Task 1 – Defining Events

```python
# Example event definitions (change column names as needed)
A = df[df['children'] > 2]
B = df[df['region'] == 'southeast']
C = df[(df['children'] >= 2) & (df['children'] <= 5)]


print('Event A size:', len(A))
print('Event B size:', len(B))
print('Event C size:', len(C))
```

## Section D: Task 2 – Calculating Basic Probability

```python
P_A = len(A) / N
P_B = len(B) / N
P_C = len(C) / N

print('P(A) =', P_A)
print('P(B) =', P_B)
print('P(C) =', P_C)
```

## Section E: Task 3 – Combined Events

```python
A_int_B = df[(df['children'] > 2) & (df['region'] == 'southeast')]
P_A_int_B = len(A_int_B) / N
print('P(A ∩ B) =', P_A_int_B)

A_union_B = df[(df['children'] > 2) | (df['region'] == 'southeast')]
P_A_union_B = len(A_union_B) / N
print('P(A ∪ B) =', P_A_union_B)

A_comp = df[df['children'] <= 2]
P_A_comp = len(A_comp) / N
print('P(Aᶜ) =', P_A_comp)

rule_value = P_A + P_B - P_A_int_B
print('Rule Verification:', rule_value)
print('Actual P(A ∪ B):', P_A_union_B)
# Verification Check
if abs(rule_value - P_A_union_B) < 1e-9:
    print("Verification: The Addition Rule holds true.")
else:
    print("Verification: Discrepancy found.")
```

## Section F: Visualization

```python
counts = [len(A), len(A_comp)]
labels = ['A: children > 2', 'Aᶜ: children ≤ 2']

plt.bar(labels, counts)
plt.title('Event A and Complement Frequency')
plt.ylabel('Count')
plt.show()
```

```python
df['region'].value_counts().plot(kind='bar')
plt.title('region Frequency')
plt.ylabel('Count')
plt.xlabel('region')
plt.show()
```
