## Part 2: Frequency Distribution Table for Age

Here, we will select a column and construct a frequency distribution table.

### Column Selection

**Justification:**
I chose the 'Age' column because it is a key demographic variable, and understanding its distribution can provide insights into the age demographics of the subjects in this dataset.

```python
# Select the column for analysis
column_to_analyze = 'age' # <-- Replace 'column_name' with the name of the column you chose

freq_table = pd.DataFrame(df[column_to_analyze].value_counts()).reset_index()
freq_table.columns = ['Class/Category', 'Frequency (f)']
freq_table = freq_table.sort_values(by='Class/Category').reset_index(drop=True)

# Step 2: Calculate Relative Frequency (rf)
total_count = freq_table['Frequency (f)'].sum()
freq_table['Relative Frequency (rf)'] = freq_table['Frequency (f)'] / total_count

# Step 3: Calculate Cumulative Frequency (cf)
freq_table['Cumulative Frequency (cf)'] = freq_table['Frequency (f)'].cumsum()

# Step 4: Calculate Relative Cumulative Frequency (rcf)
freq_table['Relative Cumulative Frequency (rcf)'] = freq_table['Relative Frequency (rf)'].cumsum()


# Display the frequency table
print("Frequency Distribution Table for '" + column_to_analyze + "'")
freq_table
```

## Part 3: Graphical Representation

In this section, we will visualize the data distribution using various charts.

### 3.1 Bar Chart / Histogram, Polygon and Ogive graphs for Age Column

```python

import pandas as pd
import matplotlib.pyplot as plt

# -------------------------------
# Load Dataset
# -------------------------------
df = pd.read_csv("insurance.csv")

# Column to analyze
column_to_analyze = "age"

# -------------------------------
# Build Frequency Table for Numerical Data
# -------------------------------
# You can change "bins=10" if needed
freq, bins = pd.cut(df[column_to_analyze], bins=10, retbins=True)
freq_table = freq.value_counts().sort_index().reset_index()
freq_table.columns = ["Class/Category", "Frequency (f)"]

# Calculate Cumulative Frequency
freq_table["Cumulative Frequency (cf)"] = freq_table["Frequency (f)"].cumsum()

# Convert class intervals to string for plotting
freq_table["Class/Category"] = freq_table["Class/Category"].astype(str)

# -------------------------------
# 1 Histogram (Matplotlib Only)
# -------------------------------
plt.figure(figsize=(12, 6))
plt.hist(df[column_to_analyze], bins=bins, edgecolor='black')

plt.title(f'Histogram of {column_to_analyze}', fontsize=16)
plt.xlabel(column_to_analyze, fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.xticks(rotation=45)
plt.show()

# -------------------------------
# 2 Frequency Polygon
# -------------------------------
plt.figure(figsize=(12, 6))
plt.plot(freq_table['Class/Category'], freq_table['Frequency (f)'],
         marker='o', linestyle='-')

plt.title(f'Frequency Polygon of {column_to_analyze}', fontsize=16)
plt.xlabel(column_to_analyze, fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# -------------------------------
# 3 Ogive (Cumulative Frequency Curve)
# -------------------------------
plt.figure(figsize=(12, 6))
plt.plot(freq_table['Class/Category'], freq_table['Cumulative Frequency (cf)'],
         marker='o', linestyle='-')

plt.title(f'Ogive Chart (Less Than) of {column_to_analyze}', fontsize=16)
plt.xlabel(column_to_analyze, fontsize=12)
plt.ylabel('Cumulative Frequency', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

```
